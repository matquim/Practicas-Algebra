## Concepto Teórico

\subsection{Práctica 2: Operaciones con conjuntos}

\subsubsection{Concepto teórico}

\begin{itemize}
    \item Unión: $A \cup B$
    \item Intersección: $A \cap B$
    \item Diferencia: $A - B$
\end{itemize}

\subsubsection{Ejemplo}

\begin{lstlisting}
A = {1, 2, 3}
B = {3, 4, 5}

print(A | B)   # union
print(A & B)   # interseccion
print(A - B)   # diferencia
\end{lstlisting}

\subsubsection{Ejercicios}

Sean $A = \{1,2,3,4,5\}$ y $B = \{4,5,6,7,8\}$:

\begin{enumerate}[label=\alph*)]
    \item $A \cup B$
    \item $A \cap B$
    \item $A - B$
    \item $B - A$
\end{enumerate}
