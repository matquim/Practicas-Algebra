\subsection{Práctica 4: Generación de conjuntos}

\subsubsection{Concepto teórico}

Se pueden generar conjuntos mediante comprensión de conjuntos.

\subsubsection{Ejemplo}

\begin{lstlisting}
def pares(n):
    return {2*i for i in range(1, n+1)}

print(pares(10))
\end{lstlisting}

\subsubsection{Ejemplo adicional}

\begin{lstlisting}
def impares(n):
    return {2*i - 1 for i in range(1, n+1)}

print(impares(10))
\end{lstlisting}

\subsubsection{Ejercicios}

\begin{enumerate}[label=\alph*)]
    \item Generar los primeros 20 números pares.
    \item Generar los primeros 20 números impares.
\end{enumerate}
