# Práctica 5: Producto cartesiano

## Concepto Teórico

\begin{equation}
A \times B = \{(a,b) \mid a \in A, b \in B\}
\end{equation}

## Ejemplo

~~~python
A = {1,2,3}
B = {"x","y"}

producto = {(a,b) for a in A for b in B}
print(producto)
~~~

## Ejercicios
