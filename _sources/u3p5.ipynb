# Funciones Proposicionales

````{admonition} Concepto
:class: tip
Una función proposicional es verdadera o falsa dependiendo del valor de su parámetro.
````



* Escribe una función que indique si un número es estrictamente mayor que $\pi$
*  Escribe una función que reciba dos números $a$ y $b$ e indique si $a$ es estrictamente mayor que $b$.




`````python
import math

def Mayor_que_pi(x):
  return x > math.pi

def Comparar_numeros(a,b):
  return a>b

print(Mayor_que_pi(2) )
print(Comparar_numeros(2,3))
`````

Utilizar estas bibliotecas para escribir un código que indique si ya pasamos el día miércoles o no.

`````python
from datetime import datetime
from zoneinfo import ZoneInfo

hoy = datetime.now(ZoneInfo("America/Mexico_City"))

def Compara_con_miercoles(n):
  return n>2

Compara_con_miercoles(hoy.weekday())
`````

# Conectivos Lógicos

````{admonition} Concepto
:class: tip
Para unir dos proposiciones atómicas y formar una
proposición molecular, podemos usar conectivos lógicos.
En Python disponemos de tres principales:

-   not   →  negación
-   and   →  conjunción
-   or    →  disyunción
````

A continuación podemos escribir algunas proposiciones moleculares utilizando estos conectivos.

`````python
not (3>6)
`````

`````python
x=19
(x<=20) and (x>16)
`````

$P \Rightarrow Q \equiv \neg P \lor Q$

`````python
def Implicacion(P,Q):
  return (not P) or Q

P=False
Q=True

Implicacion(P,Q)
`````

En Python, los conectivos lógicos tienen un orden de prioridad:

1. not   (mayor prioridad)
2. and
3. or    (menor prioridad)

Así, la expresión:

~~~python
not (x == 3) and (x < 4)
~~~

se evalúa como:

~~~python
(not (x == 3)) and (x < 4)
~~~

Es decir, primero se niega (x == 3), y luego
se verifica la conjunción con (x < 4).
Para evitar ambigüedades, siempre es recomendable
usar paréntesis de forma explícita.
