## Concepto Teórico

\subsubsection{Concepto teórico}

\[
A \times B = \{(a,b) \mid a \in A, b \in B\}
\]

\subsubsection{Ejemplo}

\begin{lstlisting}
A = {1,2,3}
B = {"x","y"}

producto = {(a,b) for a in A for b in B}
print(producto)
\end{lstlisting}

\subsubsection{Ejercicios}

Sean $X = \{1,2\}$, $Y = \{a,b\}$, $Z = \{x,y\}$:

\begin{enumerate}[label=\alph*)]
    \item $X \times Y$
    \item $Y \times Z$
    \item $X \times Y \times Z$
\end{enumerate}
