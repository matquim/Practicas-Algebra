# Práctica 1: El bucle $\texttt{for}$, lo básico

## Objetivo

Aprender a utilizar el bucle `for` para repetir instrucciones y generar secuencias numéricas.

## Concepto teórico

El bucle `for` permite ejecutar un bloque de código varias veces. Su estructura general es:

~~~python
for i in range(inicio, fin, paso):
    instrucciones
~~~

- \texttt{inicio}: valor inicial
- \texttt{fin}: valor final (no incluido)
- \texttt{paso}: incremento (opcional)

\textcolor{red}{\textbf{Nota:} Observe que las instrucciones que se deben ejecutar dentro del bucle están indentadas (desplazadas hacia la derecha). Esta indentación es obligatoria en Python y define el bloque de código del `for`.}

## Ejemplo 1
Contaremos del 1 al 10 usando el bucle \textit{for}
~~~python
### Imprimir los numeros del 1 al 10 con un bucle for
print("Contando del 1 al 10")
for i in range(1,11):
    print(i)
~~~


\subsubsection{ejemplo 2}

Imprimir los números pares en el intervalo $(1,20)$:
\begin{lstlisting}
for i in range(2,21,2):
    print(i)
\end{lstlisting}

Se podría realizar también de la siguiente manera, aprovechando la definición de paridad (todo número par es de la forma $2k$ con $k\in\mathbb{Z}$):

\begin{lstlisting}
for i in range(1,11):
    print(2*i)
\end{lstlisting}


\subsubsection{Ejemplo 3}
Imprimir los cuadrados y los cubos de los números naturales en el intervalo [1,10]

\begin{lstlisting}
for i in range(1,11):
    print("x =", i, "x^2 =", i**2, "x^3 =", i**3)
\end{lstlisting}
\textcolor{red}{\textbf{Nota:} Observe que los elementos escritos entre comillas se muestran literalmente en la salida del programa, es decir, aparecen exactamente como fueron escritos.}




\subsubsection{Ejemplo 4}

Imprimir la suma de los números naturales del 1 al 20, incluyendo su suma parcial.

\begin{lstlisting}
suma = 0
print("\nSumando los numeros naturales del 1 al 20")

for i in range(1,21):
    suma += i
    print("Suma parcial:", suma)

# note como el cambio en la indentacion da fin al bucle
print("La suma total es:", suma)  # Esta accion solo se ejecuta una vez, no es parte del bucle
\end{lstlisting}

\textcolor{red}{\textbf{Nota:} La variable \texttt{suma} se inicializa en 0 antes de comenzar el bucle. Esta inicialización es necesaria para poder acumular correctamente los valores en cada iteración.}

\textcolor{red}{\textbf{Nota:} Observe que al terminar la indentación del bloque \texttt{for}, el programa sale del bucle. Por ello, la última instrucción \texttt{print} se ejecuta una sola vez.}
