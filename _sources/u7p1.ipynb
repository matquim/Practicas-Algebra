# Definición de Polinomios

## Objetivo

Aprender a definir y manejar polinomios.

````{admonition} Concepto
:class: tip

Un polinomio es una expresión algebráica del tipo
\begin{equation}
P(x) = a_0 + a_1 x + a_2 x^2 + \cdots + a_n x^n
\end{equation}

Note que todas las potencias son números naturales.

El `grado` del polinomio está determinado por la potencia más alta.

Por ejemplo,
\begin{equation}
P(x) = 5 + 3 x - 2 x^2
\end{equation}
es un polinomio de grado dos.

Los polinomios también se clasifican por su número de términos, por ejemplo,
| Nombre        |    Ejemplo    |
|:-------------:|:-------------:|
|Monomio        |     $3x^2$    |
|Binomio        |     $2x+5$    |
|Trinomio       |$x^5 + 2x - 4$ |

````
