# Práctica 3: Conjunto universal y complemento

## Concepto Teórico

El conjunto universal $U$ contiene todos los elementos considerados.

El complemento se define como:
\[
A^c = U - A
\]

## Ejemplo

\begin{lstlisting}
A = {"a","b","c"}
B = {"+","x","="}
C = {1,2,3}

U = A | B | C

print("Complemento de A:", U - A)
\end{lstlisting}

## Ejercicios

\begin{enumerate}[label=\alph*)]
    \item Definir $A$ y $B$.
    \item Construir $U$.
    \item Calcular $A^c$ y $B^c$.
\end{enumerate}
