# Práctica 7: Ángulo entre vectores

## Concepto Teórico

\begin{equation}
\cos \theta = \frac{p \cdot q}{||p||\,||q||}
\end{equation}

## Ejemplo

~~~python
import math as mt

p = [1, 2, 2]
q = [-1, 0, 2]

dot = sum(p[i]*q[i] for i in range(3))
norm_p = (sum(p[i]**2 for i in range(3)))**0.5
norm_q = (sum(q[i]**2 for i in range(3)))**0.5

theta = mt.acos(dot / (norm_p * norm_q))
print(mt.degrees(theta))
~~~
