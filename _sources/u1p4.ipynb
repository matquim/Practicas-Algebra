## Concepto Teórico

\subsection{Práctica 4: Combinaciones}

\subsubsection{Concepto teórico}

Las combinaciones se calculan como:

\begin{equation}
C(n,k) = \frac{n!}{k!(n-k)!}
\end{equation}

\subsubsection{Ejercicios}

\begin{enumerate}[label=\alph*)]
    \item Calcular $C(10,2)$
    \item Calcular $C(100,4)$
    \item Calcular $C(500,3)$
\end{enumerate}
