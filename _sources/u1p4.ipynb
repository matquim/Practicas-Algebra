# Factorial

## Objetivo

Implementar el cálculo del factorial sin usar funciones predefinidas.

````{admonition} Concepto
:class: tip

El factorial de un número $n$ se define como:

\begin{equation}
n! = 1 \cdot 2 \cdot 3 \cdots n
\end{equation}
````

## Ejemplo

Utilice un `for` para calcular el factorial del número 4.

`````python
# Escriba código aquí
`````

`````python
# Calculo del factorial de 4
n = 4
factorial = 1

for i in range(1, n+1):
    factorial *= i

print("Factorial:", factorial)
`````

Ahora utilice un `for` para calcular el factorial del número 100.

`````python
# Escriba código aquí
`````

`````python
# Calculo del factorial de 100
n = 20
factorial = 1

for i in range(1, n+1):
    factorial *= i

print("Factorial:", factorial)
`````
