# Operaciones con conjuntos

````{admonition} Concepto
:class: tip

- Unión: $A \cup B$
- Intersección: $A \cap B$
- Diferencia: $A - B$
````

## Ejemplo

~~~python
A = {1, 2, 3}
B = {3, 4, 5}

print(A | B)   # union
print(A & B)   # interseccion
print(A - B)   # diferencia
~~~

## Ejercicios

Sean $A = \{1,2,3,4,5\}$ y $B = \{4,5,6,7,8\}$:

- $A \cup B$
- $A \cap B$
- $A - B$
- $B - A$

# Producto cartesiano

````{admonition} Concepto
:class: tip

\begin{equation}
A \times B = \{(a,b) \mid a \in A, b \in B\}
\end{equation}
````

## Ejemplo

~~~python
A = {1,2,3}
B = {"x","y"}

producto = {(a,b) for a in A for b in B}
print(producto)
~~~

## Ejercicios

Sean $X = \{1,2\}$, $Y = \{a,b\}$, $Z = \{x,y\}$:

- $X \times Y$
- $Y \times Z$
- $X \times Y \times Z$
