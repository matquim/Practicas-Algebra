# Operaciones con conjuntos

````{admonition} Concepto
:class: tip

- Unión: $A \cup B$
- Intersección: $A \cap B$
- Diferencia: $A - B$
````

## Ejemplo

~~~python
A = {1, 2, 3}
B = {3, 4, 5}

print(A | B)   # union
print(A & B)   # interseccion
print(A - B)   # diferencia
~~~

## Ejercicios

Sean $A = \{1,2,3,4,5\}$ y $B = \{4,5,6,7,8\}$:

- $A \cup B$
- $A \cap B$
- $A - B$
- $B - A$
