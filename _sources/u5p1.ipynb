# Práctica 1: Definición de conjuntos en Python

## Objetivo

Aprender a definir conjuntos y manipular sus elementos en Python.

## Concepto Teórico

Un conjunto es una colección de elementos sin orden y sin repetición.

## Ejemplo

~~~python
# Conjunto de letras
A = {"a", "b", "c"}

# Conjunto de numeros
B = {1, 2, 3, 4}
~~~
