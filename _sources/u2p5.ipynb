# Práctica 5: Producto punto

## Concepto Teórico

\begin{equation}
a \cdot b = \sum_i a_i b_i
\end{equation}

## Ejemplo

~~~python
a = [2, -1, 5]
b = [3, 2, -4]

dot = sum(a[i] * b[i] for i in range(3))
print(dot)
~~~
