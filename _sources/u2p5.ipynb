## Concepto Teórico

\subsubsection{Concepto teórico}

\begin{equation}
a \cdot b = \sum_i a_i b_i
\end{equation}

\subsubsection{Ejemplo}

\begin{lstlisting}
a = [2, -1, 5]
b = [3, 2, -4]

dot = sum(a[i] * b[i] for i in range(3))
print(dot)
\end{lstlisting}

\subsubsection{Ejercicios}

Sean $u=(1,2,3)$, $v=(2,4,6)$, $w=(-1,0,1)$:

\begin{enumerate}[label=\alph*)]
    \item $u \cdot v$
    \item $v \cdot w$
    \item $w \cdot (u+v)$
\end{enumerate}
