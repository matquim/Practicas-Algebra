### Ejemplo de Base y dimensión en $\mathbb{R}^3$

1. Verificar si los vectores $\mathbf{u}$, $\mathbf{v}$ y $\mathbf{w}$ forman una base de $\mathbb{R}^3$

$$
\mathbf{u}=
\begin{pmatrix}
1\\
2\\
0
\end{pmatrix},
\qquad
\mathbf{v}=
\begin{pmatrix}
0\\
1\\
1
\end{pmatrix},
\qquad
\mathbf{w}=
\begin{pmatrix}
2\\
0\\
1
\end{pmatrix}
$$

Usando Python podemos hacer los siguiente

* Construir una matriz $A$ cuyas columnas sean los vectores $\mathbf{u}$, $\mathbf{v}$ y $\mathbf{w}$

* Usar Python para calcular el determinante de $A$ y determinar si los vectores efectivamente forman una base del espacio $\mathbb{R}^3$


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
u =  np.array([1, 2, 0])
v =  np.array([0, 1, 1])
w =  np.array([2, 0, 1])

A = np.column_stack((u, v, w))

det_A = np.linalg.det(A)

print(det_A)
```

$\mathbf{u}$, $\mathbf{v}$ y $\mathbf{w}$ son linealmente independientes y por lo tanto forman una base de $\mathbb{R}^3$



2. Sea $
\mathbf{p}=
\begin{pmatrix}
3\\
4\\
2
\end{pmatrix}
$

Exprese el vector $\mathbf{p}$ como una combinación lineal de los vectores $\mathbf{u}$, $\mathbf{v}$ y $\mathbf{w}$, es decir

$ \mathbf{p} = \alpha_1\mathbf{u} + \alpha_2\mathbf{v} + \alpha_3\mathbf{w}$


```python
d = np.array([3, 4, 2])

coeficientes =  np.linalg.solve(A, d)

#print(coeficientes)
print(f" p = {coeficientes[0]}u + {coeficientes[1]}v {coeficientes[2]}w")
```

### Ejemplos de espacios generados

1. Considere el vector

$$
\mathbf{v_1}=
\begin{pmatrix}
2\\
1
\end{pmatrix}
$$

El espacio generado por $\mathbf{v_1}$ está dado por

$$
\operatorname{gen}\left\{\mathbf{v_1}
\right\}
=
\left\{\mathbf{v}: \mathbf{v} = a\mathbf{v_1}, a\in\mathbb{R}
\right\}
$$


```python
v1 = np.array([2, 1])

a_values = np.linspace(-3, 3, 500)

x = []
y = []

for a in a_values:
    v = a*v1
    x.append(v[0])
    y.append(v[1])


plt.scatter(x, y, s=2, color="gray")
plt.quiver(0, 0, v1[0], v1[1],
           angles="xy",
           scale_units="xy",
           scale=1,
           color="green")

plt.text(v1[0] + 0.1 , v1[1] - 0.5 , r'(2,1)',
         fontsize=12, color='green')

plt.axis("equal")
plt.grid()
plt.show()
```

2. Considere los vectores

$$
\mathbf{w_1}=
\begin{pmatrix}
1\\
0
\end{pmatrix},
\qquad
\mathbf{w_2}=
\begin{pmatrix}
0\\
1
\end{pmatrix}
$$

El espacio generado por $\mathbf{w_1}$ y $\mathbf{w_2}$ está dado por

$$
\operatorname{gen}\left\{\mathbf{w_1},\mathbf{w_2}\right\}
=
\left\{
\mathbf{w}: \mathbf{w}=a\mathbf{w_1}+b\mathbf{w_2},
\;a,b\in\mathbb{R}
\right\},
$$

en donde

$$
a\mathbf{w_1}+b\mathbf{w_2}
=
a
\begin{pmatrix}
1\\
0
\end{pmatrix}
+
b
\begin{pmatrix}
0\\
1
\end{pmatrix}
=
\begin{pmatrix}
a\\
b
\end{pmatrix}
$$



```python
w1 = np.array([1, 0])
w2 = np.array([0, 1])

x = []
y = []

for a in np.linspace(-3, 3, 100):
    for b in np.linspace(-3, 3, 100):
        w = a*w1 + b*w2
        x.append(w[0])
        y.append(w[1])

plt.scatter(x, y, color="gray", s=1)

plt.quiver(0, 0, w1[0], w1[1],
           angles="xy",
           scale_units="xy",
           scale=1,
           color="red")

plt.quiver(0, 0, w2[0], w2[1],
           angles="xy",
           scale_units="xy",
           scale=1,
           color="blue")

plt.axis("equal")
plt.grid()
plt.show()



```

### Representación gráfica de complejos

```python
import matplotlib.pyplot as plt

# Números complejos
z1 = 2 + 3j
z2 = 1 - 4j

# Coordenadas de los puntos
x1 = z1.real
y1 = z1.imag

x2 = z2.real
y2 = z2.imag

plt.figure()

# Punto correspondiente al número complejo z1
plt.scatter(x1, y1, color='blue', s=100)

# Vector desde el origen
plt.quiver(
    0, 0, x1, y1,
    angles='xy',
    scale_units='xy',
    scale=1,
    color='blue',
    headwidth=1,
    headlength=0,
    headaxislength=1
)

# Etiqueta de z1
plt.text(x1 + 0.2, y1, r'$2+3i$',
         fontsize=12, color='blue')

# Punto correspondiente al número complejo z2
plt.scatter(x2, y2, color='green', s=100)

# Vector desde el origen
plt.quiver(
    0, 0, x2, y2,
    angles='xy',
    scale_units='xy',
    scale=1,
    color='green',
    headwidth=1,
    headlength=0,
    headaxislength=1
)

# Etiqueta de z2
plt.text(x2 + 0.2, y2, r'$1-4i$',
         fontsize=12, color='green')

# Ejes
plt.axhline(0, color='black')
plt.axvline(0, color='black')

plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
plt.xlim(-7, 7)
plt.ylim(-7, 7)

plt.grid()
plt.show()
```

### Raíces de un número complejo
Si

$$
z=r(\cos\theta+i\sin\theta)
$$

entonces las $n$ raíces de $z$ están dadas por

$$
z_k=\sqrt[n]{r}\left(\cos\left(\frac{\theta+2k\pi}{n}\right)
+i\sin\left(\frac{\theta+2k\pi}{n}\right)\right)
\qquad
k=0,1,\ldots,n-1.
$$


Ejemplo

Determine las raíces cúbicas de la unidad

$$
z=1
$$

i.e. todos los números complejos $w$ tales que

$$
w^3=1
$$




```python
# número complejo

z = 1 + 0j

r = np.abs(z)
theta = np.angle(z)

n = 3

for k in range(n):
    phi = (theta + 2*np.pi*k)/n
    rho = r**(1/n)

    z_k = rho*(np.cos(phi) + 1j*np.sin(phi))

    print(f"Raíz {k + 1}:")
    print("Forma rectangular:", z_k)
    print(f"Forma polar: {rho}(Cos({phi:.3f} + i Sin({phi:.3f})))")
```
