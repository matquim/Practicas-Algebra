## Concepto Teórico

\subsubsection{Concepto teórico}

El producto cruz en $\mathbb{R}^3$ es:

\[
p \times q =
\begin{pmatrix}
p_y q_z - p_z q_y \\
p_z q_x - p_x q_z \\
p_x q_y - p_y q_x
\end{pmatrix}
\]

\subsubsection{Ejemplo}

\begin{lstlisting}
p = [1, 0, -2]
q = [3, -1, 1]

cross = [
    p[1]*q[2] - p[2]*q[1],
    p[2]*q[0] - p[0]*q[2],
    p[0]*q[1] - p[1]*q[0]
]

print(cross)
\end{lstlisting}
