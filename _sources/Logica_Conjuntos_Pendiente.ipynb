# En Python, los conectivos lógicos tienen un orden de prioridad:
#
#   1. not   (mayor prioridad)
#   2. and
#   3. or    (menor prioridad)
#
# Así, la expresión:
#
#   not (x == 3) and (x < 4)
#
# se evalúa como:
#
#   (not (x == 3)) and (x < 4)
#
# Es decir, primero se niega (x == 3), y luego
# se verifica la conjunción con (x < 4).
# Para evitar ambigüedades, siempre es recomendable
# usar paréntesis de forma explícita.

# -----------------------------------------------------------
# 4. TABLAS DE VERDAD
# Usaremos estos conectivos para construir tablas de verdad.
# Cada renglón corresponde a un juego de valores de verdad
# de las proposiciones atómicas.
#
# Para dos proposiciones P y Q, hay 4 combinaciones posibles:
#
#   P      Q
#  ────  ────
#  True  True
#  True  False
#  False True
#  False False
#
# Construyamos la tabla de verdad de la conjunción (and):
# -----------------------------------------------------------
```python
P = True;  Q = True;  print(P, Q, "→", P and Q)
P = True;  Q = False; print(P, Q, "→", P and Q)
P = False; Q = True;  print(P, Q, "→", P and Q)
P = False; Q = False; print(P, Q, "→", P and Q)
```
# Nota: en este ejemplo construimos cada renglón "a mano".
# ¿Cómo podemos mejorar ese proceso?

````python
print("P\tQ\tP y Q\tP o Q\t P => Q\t P <=> Q ")

for P in [True, False]:
  for Q in [True, False]:
    print(P, "\t", Q, "\t", P and Q, "\t", P or Q, "\t", Implicacion(P,Q), "\t", P == Q)
````

# 5. EQUIVALENCIAS LÓGICAS
# Dos proposiciones son equivalentes si tienen el mismo valor de verdad para cualquier combinación de valores de verdad de las proposiciones que las componen.
#
# Verificaremos las siguientes equivalencias:
#
#  1. Ley de De Morgan:
#       $\neg(P \wedge Q) \equiv \neg P \vee \neg Q$
#
#  2. Distributividad:
#       $P \wedge (Q \vee R) \equiv (P \wedge Q) \vee (P \wedge R)$
#
#  3. Negación de la implicación:
#       $\neg(P \Rightarrow Q) \equiv P \wedge \neg Q$
#
#  4. Contrapuesta:
#       $P \Rightarrow Q \equiv \neg Q \Rightarrow \neg P$
#
#  5. La implicación no es conmutativa:
#       $P \Rightarrow Q \not\equiv Q \Rightarrow P$
#
# Nota: la implicación ($\Rightarrow$) no existe como tal en Python.
# Usaremos la equivalencia que podemos definir como funcion:
#       $P \Rightarrow Q \equiv \neg P \vee Q$
### -----------------------------------------------------------

````python
print("P \tQ \tneg(P y Q) \tneg P o neg Q")
for P in [True, False]:
  for Q in [True, False]:
    print(P ,"\t" ,Q, "\t", not(P and Q), "\t", not P or not Q)
````

````python
print("P\tQ\tR\t Py(QoR)\t\t(PyQ) o (PyR)")


for P in [True, False]:
  for Q in [True, False]:
    for R in [True, False]:
      print(P, "\t", Q, "\t", R, "\t", P and (Q or R), "\t\t", (P and Q) or (P and R))
````

# 6. LEYES VÁLIDAS DE INFERENCIA
# Una ley de inferencia nos permite concluir una proposición a partir de otras proposiciones dadas. Estas reglas son siempre válidas, es decir, son tautologías: su valor de verdad es True para cualquier combinación de valores de verdad de las proposiciones que las componen.
#
# Verificaremos las siguientes leyes:
#
#  1. Modus ponens:
#       $(P \land (P \Rightarrow Q)) \Rightarrow Q$
#
#  2. Modus tollendo ponens:
#       $((\neg P) \land (P \vee Q)) \Rightarrow Q$
#
#  3. Modus tollendo tollens:
#       $((\neg Q) \land (P \Rightarrow Q)) \Rightarrow (\neg P)$
#
#  4. Silogismo:
#       $((P \Rightarrow Q) \land (Q \Rightarrow R))    \Rightarrow (P \Rightarrow R)$



````python
print("P\tQ\tR\tSilogismo")

for P in [True, False]:
  for Q in [True, False]:
    for R in [True, False]:
      A=Implicacion(P,Q)
      B=Implicacion(Q,R)
      C=Implicacion(P,R)
      Silogismo=Implicacion(A and B, C)
      print(P, "\t", Q, "\t", R, "\t", Silogismo )
````
