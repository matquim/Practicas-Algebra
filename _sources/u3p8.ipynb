# Práctica 8: Aplicaciones

## Ejercicios

\begin{enumerate}[label=\alph*)]
    \item Construir:
    \[
    X = \{x \in \mathbb{N} \mid 1 \le x \le 10\}
    \]
    \item Definir:
    \[
    Y = \{x^2 \mid x \in X\}
    \]
    \item Definir el conjunto de números primos en $X$.
    \item Calcular:
    \begin{itemize}
        \item $X \cap Y$
        \item $Y - Z$
        \item $(X \cup Y) \cap Z$
    \end{itemize}
\end{enumerate}
