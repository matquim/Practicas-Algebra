# Práctica 1: Representación de vectores

## Objetivo

Aprender a representar vectores en Python y acceder a sus componentes.

## Concepto Teórico

Un vector en $\mathbb{R}^n$ se representa como una lista de $n$ componentes.

## Ejemplo

~~~python
# Vector en R2
vec1 = [1, 2]
print(vec1)

# Vector en R3
vec2 = [1.0, -2, 3]
print(vec2)
~~~

## Acceso a componentes

~~~python
vec = [1, -2, 3]

print(vec[0])  # x
print(vec[1])  # y
print(vec[2])  # z
~~~
