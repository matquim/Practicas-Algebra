\subsection{Práctica 2: Aplicaciòn al càlculo de sumas y series}

\subsubsection{Objetivo}
Calcular sumas utilizando programación y compararlas con fórmulas analíticas.

\subsubsection{Concepto teórico}

Se sabe que el resultado de la suma de los primeros n números naturales es: 

\begin{equation}
\sum_{i=1}^{n} i = \frac{n(n+1)}{2}
\end{equation}

\subsubsection{Ejemplo 1}

Implemente un código que calcule esta suma y utilícelo para calcular los valores de la suma cuando $n=20,100,10000$. Compare con el resultado analítico de la ecuación anterior.

\begin{lstlisting}
# Comparacion sin usar funciones
for n in [20, 100, 10000]:
    suma = 0
    for i in range(1,n+1):
        suma += i

    suma_analitica = n*(n+1)//2

    print("n =", n)
    print("Suma numerica:", suma)
    print("Suma analitica:", suma_analitica)
    print("Diferencia:", suma - suma_analitica)
    print()
\end{lstlisting}

\textcolor{red}{\textbf{Nota:} La expresión analítica de la suma de los primeros $n$ números naturales es $S = \frac{n(n+1)}{2}$.}


\subsubsection{Ejemplo 1 más complejo: usando funciones}

\begin{lstlisting}
# Funcion para la suma numerica
def suma_numerica(n):
    suma = 0
    for i in range(1,n+1):
        suma += i
    return suma

# Funcion para la solucion analitica
def suma_analitica(n):
    return n*(n+1)//2

# Comparacion
for n in [20, 100, 10000]:
    print("n =", n)
    print("Suma numerica:", suma_numerica(n))
    print("Suma analitica:", suma_analitica(n))
    print("Diferencia:", suma_numerica(n) - suma_analitica(n))
    print()
\end{lstlisting}

\textcolor{red}{\textbf{Nota:} El uso de funciones permite reutilizar código y hacer el programa más claro y modular.}
