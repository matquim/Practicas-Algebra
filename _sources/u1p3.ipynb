\subsection{Práctica 3: Factorial}

\subsubsection{Objetivo}
Implementar el cálculo del factorial sin usar funciones predefinidas.

\subsubsection{Concepto teórico}

El factorial de un número $n$ se define como:

\begin{equation}
n! = 1 \cdot 2 \cdot 3 \cdots n
\end{equation}

\subsubsection{Ejemplo}

\begin{lstlisting}
# Calculo del factorial de 4
n = 4
factorial = 1

for i in range(1, n+1):
    factorial *= i

print("Factorial:", factorial)
\end{lstlisting}

\subsubsection{Ejercicios propuestos}

\begin{enumerate}[label=\alph*)]
    \item Calcular el factorial de diferentes valores de $n$.
    \item Comparar con la función \texttt{math.factorial()}.
    \item Desarrolle un código que imprima las tablas de multiplicar del 1 al 10.
\end{enumerate}
