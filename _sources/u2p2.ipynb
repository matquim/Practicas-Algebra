# Práctica 2: Suma de vectores

## Concepto Teórico

La suma de vectores se realiza componente a componente.

## Ejemplo

~~~python
u = [1, -2, 3]
v = [4, 0, -1]

suma = [u[i] + v[i] for i in range(3)]
print("u + v =", suma)
~~~

## Ejercicios

- Calcular $u + v$ para distintos vectores.
- Implementar la suma sin comprensión de listas.
