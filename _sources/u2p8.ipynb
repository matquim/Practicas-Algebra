## Visualización (opcional)

~~~python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

a = np.array([2,3,1])
b = np.array([-1,2,4])
origin = np.array([0,0,0])

ax.quiver(*origin, *a)
ax.quiver(*origin, *b)

plt.show()
~~~
