# Práctica 4: Generación de conjuntos

## Concepto Teórico

Se pueden generar conjuntos mediante comprensión de conjuntos.

## Ejemplo

~~~python
def pares(n):
    return {2*i for i in range(1, n+1)}

print(pares(10))
~~~

## Ejemplo adicional

~~~python
def impares(n):
    return {2*i - 1 for i in range(1, n+1)}

print(impares(10))
~~~

## Ejercicios

- Generar los primeros 20 números pares.
- Generar los primeros 20 números impares.
