# Práctica 6: Subconjuntos

## Objetivo

## Concepto Teórico



\subsubsection{Concepto teórico}

Un conjunto $A$ es subconjunto de $B$ si todos sus elementos están en $B$.

\subsubsection{Ejemplo}

\begin{lstlisting}
A = {1,2}
B = {1,2,3,4}

print(A.issubset(B))
\end{lstlisting}

\subsubsection{Ejercicios}

\begin{enumerate}[label=\alph*)]
    \item Verificar si un conjunto es subconjunto de otro.
\end{enumerate}
