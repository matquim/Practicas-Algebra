# Práctica 6: Subconjuntos

## Concepto Teórico

Un conjunto $A$ es subconjunto de $B$ si todos sus elementos están en $B$.

## Ejemplo

\begin{lstlisting}
A = {1,2}
B = {1,2,3,4}

print(A.issubset(B))
\end{lstlisting}

## Ejercicios

\begin{enumerate}[label=\alph*)]
    \item Verificar si un conjunto es subconjunto de otro.
\end{enumerate}
