\subsection{Práctica 5: Números primos}

\subsubsection{Ejemplo}

\begin{lstlisting}
N = 30
for num in range(2, N + 1):
    es_primo = True
    for divisor in range(2, num):
        if num % divisor == 0:
            es_primo = False
            break
    if es_primo:
        print(num)
\end{lstlisting}
