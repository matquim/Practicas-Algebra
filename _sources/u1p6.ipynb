## Ejemplo

~~~python
a = 1
N = 10

for n in range(1, N + 1):
    print(f"a_{n} = {a}")
    a = (2 * a) + 3
~~~




## Ejercicios
    
- Considere una población bacteriana en la que cada generación cumple la siguiente regla: por cada 2 individuos, se generan 4 nuevos individuos.

- Suponga que la población inicial es de $a_0 = 2$ bacterias.
  - Escriba la relación de recurrencia que describe la evolución de la población.
  - Implemente un código que calcule el número de bacterias en las primeras $N = 10$ generaciones.
  -  Imprima los valores de la población en cada generación.

- Modifique el modelo anterior considerando que cada individuo sobrevive únicamente durante dos generaciones y posteriormente muere.
  - Proponga una relación de recurrencia que tenga en cuenta tanto la reproducción como la muerte de los individuos.
  - **Indicación:** La población en la generación $n+1$ depende de las generaciones anteriores.
  - Implemente un código que modele esta nueva dinámica poblacional.
  - Imprima la población en cada generación para $N = 10$.
  - Compare el comportamiento de este modelo con el modelo del inciso anterior.

~~~python
    # Modelo 1: crecimiento sin muerte
a = 2   # población inicial
N = 10
print("Modelo sin muerte:")
for n in range(N):
    print(f"Generación {n}: {a}")
    a = 3 * a
~~~

~~~python
    # Modelo 2: cada individuo vive 2 generaciones
a_prev = 2   # a_0
a = 6        # a_1 = 3 * a_0
N = 10
print("\nModelo con vida de 2 generaciones:")
print(f"Generación 0: {a_prev}")
print(f"Generación 1: {a}")
for n in range(1, N):
    a_next = 3 * a - a_prev
    print(f"Generación {n+1}: {a_next}")
    a_prev = a
    a = a_next
~~~
