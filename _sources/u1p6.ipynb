# Práctica 6: Números primos

````{admonition} Concepto
:class: tip

Un número primo es un número natural con solo dos divisores, el número uno y el propio número.

Por ejemplo, el número 5 es primo, porque sus únicos divisores son $\{1, 5\}$.

El 4 no es un número primo, porque tiene como divisores $\{1, 2 y 4\}$.
````

## Ejercicio.

Desarrolle un código que evalúe si determinado número es un número primo.

`````python
# Escriba código aquí
`````

`````python
N = 30
for num in range(2, N + 1):
    es_primo = True
    for divisor in range(2, num):
        if num % divisor == 0:
            es_primo = False
            break
    if es_primo:
        print(num)
`````
