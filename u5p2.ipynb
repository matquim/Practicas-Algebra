### Ejemplos de espacios generados

1. Considere el vector

$$
\mathbf{v_1}=
\begin{pmatrix}
2\\
1
\end{pmatrix}
$$

El espacio generado por $\mathbf{v_1}$ está dado por

$$
\operatorname{gen}\left\{\mathbf{v_1}
\right\}
=
\left\{\mathbf{v}: \mathbf{v} = a\mathbf{v_1}, a\in\mathbb{R}
\right\}
$$


2. Considere los vectores

$$
\mathbf{w_1}=
\begin{pmatrix}
1\\
0
\end{pmatrix},
\qquad
\mathbf{w_2}=
\begin{pmatrix}
0\\
1
\end{pmatrix}
$$

El espacio generado por $\mathbf{w_1}$ y $\mathbf{w_2}$ está dado por

$$
\operatorname{gen}\left\{\mathbf{w_1},\mathbf{w_2}\right\}
=
\left\{
\mathbf{w}: \mathbf{w}=a\mathbf{w_1}+b\mathbf{w_2},
\;a,b\in\mathbb{R}
\right\},
$$

en donde

$$
a\mathbf{w_1}+b\mathbf{w_2}
=
a
\begin{pmatrix}
1\\
0
\end{pmatrix}
+
b
\begin{pmatrix}
0\\
1
\end{pmatrix}
=
\begin{pmatrix}
a\\
b
\end{pmatrix}
$$
