## Ejercicios

- Construir:
    \[
    X = \{x \in \mathbb{N} \mid 1 \le x \le 10\}
    \]
- Definir:
    \[
    Y = \{x^2 \mid x \in X\}
    \]
- Definir el conjunto de números primos en $X$.
- Calcular:
  - $X \cap Y$
  - $Y - Z$
  - $(X \cup Y) \cap Z$
