## Concepto Teórico



\subsubsection{Objetivo}
Aprender a representar vectores en Python y acceder a sus componentes.

\subsubsection{Concepto teórico}

Un vector en $\mathbb{R}^n$ se representa como una lista de $n$ componentes.

\subsubsection{Ejemplo}

\begin{lstlisting}
# Vector en R2
vec1 = [1, 2]
print(vec1)

# Vector en R3
vec2 = [1.0, -2, 3]
print(vec2)
\end{lstlisting}

\subsubsection{Acceso a componentes}

\begin{lstlisting}
vec = [1, -2, 3]

print(vec[0])  # x
print(vec[1])  # y
print(vec[2])  # z
\end{lstlisting}
