# El bucle $\texttt{for}$, lo básico

## Objetivo

Aprender a utilizar el bucle `for` para repetir instrucciones y generar secuencias numéricas.

````{admonition} Concepto
:class: tip

Además de variables, Python contiene varias estructuras especiales. Una de ellas es el bucle `for` permite ejecutar un bloque de código varias veces. Su estructura general es:

~~~python
for i in range(inicio, fin, paso):
    instrucciones
~~~

- `inicio`: valor inicial
- `fin`: valor final (no incluido)
- `paso`: incremento (opcional)

**Nota** Observe que las instrucciones que se deben ejecutar dentro del bucle están indentadas (desplazadas hacia la derecha). Esta indentación de cuatro espacios es obligatoria en Python y define las instrucciones que son controladas por el `for`.
````

## Ejemplo 3
Imprimir los cuadrados y los cubos de los números naturales en el intervalo [1,10]

`````python
# Agregue código aquí
`````

`````python
for i in range(1,11):
    print("x =", i, "x^2 =", i**2, "x^3 =", i**3)
`````

**Nota:** Observe que los elementos escritos entre comillas se muestran literalmente en la salida del programa, es decir, aparecen exactamente como fueron escritos.

## Ejemplo 4

Imprimir la suma de los números naturales del 1 al 20, incluyendo su suma parcial.

`````python
# Agregue código aquí
`````

`````python
suma = 0
print("\nSumando los numeros naturales del 1 al 20")

for i in range(1,21):
    suma += i
    print("Suma parcial:", suma)

# note como el cambio en la indentacion da fin al bucle
print("La suma total es:", suma)  # Esta accion solo se ejecuta una vez, no es parte del bucle
`````

**Nota:** La variable `suma` se inicializa en 0 antes de comenzar el bucle. Esta inicialización es necesaria para poder acumular correctamente los valores en cada iteración.
**Nota:** Observe que al terminar la indentación del bloque `for`, el programa sale del bucle. Por ello, la última instrucción `print` se ejecuta una sola vez.
