# Práctica 3: Combinaciones lineales

## Concepto Teórico

Una combinación lineal es de la forma:

\begin{equation}
\alpha u + \beta v
\end{equation}

## Ejemplo

~~~python
u = [1, -2, 3]
v = [4, 0, -1]

comb = [2*u[i] - 3*v[i] for i in range(3)]
print(comb)
~~~
