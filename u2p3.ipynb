\subsection{Práctica 3: Combinaciones lineales}

\subsubsection{Concepto teórico}

Una combinación lineal es de la forma:
\[
\alpha u + \beta v
\]

\subsubsection{Ejemplo}

\begin{lstlisting}
u = [1, -2, 3]
v = [4, 0, -1]

comb = [2*u[i] - 3*v[i] for i in range(3)]
print(comb)
\end{lstlisting}

\subsubsection{Ejercicios}

Sean $u=(1,3,4)$, $v=(0,1,-1)$, $w=(0,1,-0.5)$:

\begin{enumerate}[label=\alph*)]
    \item $5u + 3v + 2w$
    \item $u + 3v - 4w$
    \item $-2u - v + 7w$
\end{enumerate}
