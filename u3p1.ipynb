\subsection{Práctica 1: Definición de conjuntos en Python}

\subsubsection{Objetivo}
Aprender a definir conjuntos y manipular sus elementos en Python.

\subsubsection{Concepto teórico}

Un conjunto es una colección de elementos sin orden y sin repetición.

\subsubsection{Ejemplo}

\begin{lstlisting}
# Conjunto de letras
A = {"a", "b", "c"}

# Conjunto de numeros
B = {1, 2, 3, 4}
\end{lstlisting}

\subsubsection{Ejercicios}

\begin{enumerate}[label=\alph*)]
    \item Definir dos conjuntos numéricos.
    \item Verificar que no se repitan elementos.
\end{enumerate}
