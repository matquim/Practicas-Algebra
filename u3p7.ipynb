\subsection{Práctica 7: Relaciones y funciones}

\subsubsection{Concepto teórico}

Una relación es un conjunto de pares ordenados.

Es función si cada elemento del dominio tiene una única imagen.

\subsubsection{Ejemplo}

\begin{lstlisting}
relacion = {(1,2),(2,4),(3,6)}

dominio = {x for (x,y) in relacion}

es_funcion = len(relacion) == len(dominio)
print(es_funcion)
\end{lstlisting}

\subsubsection{Ejercicios}

\begin{enumerate}[label=\alph*)]
    \item Verificar si una relación es función.
    \item Analizar si es biyectiva.
\end{enumerate}
