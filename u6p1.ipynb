# Práctica 1: Definición de números complejos

## Objetivo

Aprender a definir conjuntos y manipular números complejos.

````{admonition} Concepto
:class: tip

Los números complejos son aquellos que tienen una parte real y una parte imaginaria, por ejemplo:

$2 + 4i$

Por convención, en Python, la parte imaginaria se representa con la letra $j$, tal que un número complejo en Python se representa como:
~~~python
2 + 4j
~~~
````
